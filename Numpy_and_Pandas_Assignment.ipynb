```
train_data = pd.DataFrame({"State" : ["A","B","C","A","A","B","A","A","C","C"],
                          "Class" : ["0","1","1","0","1","1","0","1","1","0"]
                         })
print(train_data)
print(50*"*")

def change_val(x):
    if x == 'A':
        return 3/5
    elif x == 'B':
        return 0/2
    elif x == 'C':
        return 1/3
    else:
        return x

def change_val_2(x):
    return 1-x
    
    
encoded_train_data = pd.DataFrame([])    
encoded_train_data['State_0'] = train_data['State'].map(change_val)
encoded_train_data['State_1'] = encoded_train_data['State_0'].map(change_val_2)
encoded_train_data
```
